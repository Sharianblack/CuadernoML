````python
¬°De una, mijo! Aqu√≠ tienes la **"Plantilla Maestra"**. Este c√≥digo combina **todo** lo que hemos visto: carga de datos, limpieza, subplots, regresi√≥n, personalizaci√≥n y anotaciones.

Gu√°rdala bien porque si ma√±ana te piden cualquier cosa, seguro puedes copiar un pedazo de aqu√≠.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. CONFIGURACI√ìN INICIAL (Est√©tica)
sns.set_theme(style="whitegrid") # Fondo limpio con cuadr√≠cula
sns.set_context("notebook")      # Tama√±o de letra est√°ndar

# 2. CARGA Y LIMPIEZA DE DATOS
# Cargamos desde la URL (puedes cambiarla por tu archivo .csv local)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/planets.csv"
df = pd.read_csv(url)

# Limpieza r√°pida: eliminamos filas con nulos para evitar errores en regresiones
df_clean = df.dropna()

# ---------------------------------------------------------
# 3. CREACI√ìN DEL LIENZO (SUBPLOTS)
# Creamos una figura con 2 filas y 2 columnas
# ---------------------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Plantilla Maestra: An√°lisis Estad√≠stico Completo", fontsize=18)

# --- GR√ÅFICO 1: Distribuci√≥n y Outliers (Boxplot) ---
sns.boxplot(data=df_clean, x='distance', ax=axes[0,0], color='skyblue')
axes[0,0].set_title("1. Detecci√≥n de Outliers (Distancia)")
axes[0,0].set_xlabel("Distancia (A√±os Luz)")

# --- GR√ÅFICO 2: Comparaci√≥n de Categor√≠as (Barplot) ---
# Mostramos el promedio de distancia por cada m√©todo de descubrimiento
top_5 = df_clean['method'].value_counts().nlargest(5).index
df_top = df_clean[df_clean['method'].isin(top_5)]

sns.barplot(data=df_top, x='method', y='distance', ax=axes[0,1], palette='viridis')
axes[0,1].set_title("2. Promedio de Distancia por M√©todo")
axes[0,1].tick_params(axis='x', rotation=30) # Rotar nombres para que no se pisen

# --- GR√ÅFICO 3: Relaci√≥n y Regresi√≥n (Scatter + RegPlot) ---
# Comparamos A√±o vs Distancia con una regresi√≥n cuadr√°tica (orden 2)
sns.regplot(data=df_top, x='year', y='distance', order=2, ax=axes[1,0], 
            scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
axes[1,0].set_title("3. Tendencia: A√±o vs Distancia (Orden 2)")

# --- GR√ÅFICO 4: Densidad Cuantitativa (KDE Plot) ---
sns.kdeplot(data=df_clean, x='year', y='distance', fill=True, cmap='magma', ax=axes[1,1])
axes[1,1].set_title("4. Densidad de Descubrimientos")

# ---------------------------------------------------------
# 4. ANOTACI√ìN FINAL (Flecha y texto)
# Se√±alamos algo importante en el gr√°fico de regresi√≥n
# ---------------------------------------------------------
axes[1,0].annotate('Crecimiento exponencial', 
                   xy=(2010, 1000), 
                   xytext=(2002, 1500),
                   arrowprops=dict(facecolor='black', shrink=0.05))

# 5. AJUSTES FINALES Y GUARDADO
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Deja espacio para el supert√≠tulo
# plt.savefig('mi_examen_final.png', dpi=300) # Opcional: guardar imagen
plt.show()

# 6. EXTRA: Matriz de Correlaci√≥n (Se suele pedir aparte)
plt.figure(figsize=(8,6))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlaci√≥n de Variables")
plt.show()

```

### üöÄ ¬øPor qu√© esta plantilla es "nivel pro"?

1. **`palette='viridis'` o `'magma'**`: Son los colores que usan los cient√≠ficos porque son f√°ciles de leer.
2. **`order=2`**: Te ayuda a demostrar que sabes que los datos no siempre son una l√≠nea recta.
3. **`fmt=".2f"`**: En el heatmap, hace que los n√∫meros solo tengan 2 decimales para que no se vea desordenado.
4. **`reset_index()` o `isin()**`: T√©cnicas de Pandas para filtrar datos r√°pido.

¬°Mucha suerte ma√±ana, pana! Si te sale una pregunta de teor√≠a sobre **"¬øCu√°ndo usar Matplotlib vs Seaborn?"**, acu√©rdate:

* **Seaborn**: Para cuando quieres ver la **relaci√≥n estad√≠stica** y la **densidad** de forma r√°pida y bonita.
* **Matplotlib**: Para cuando quieres **control total** sobre los ejes, poner flechas exactas o hacer un dise√±o muy espec√≠fico.

¬øHay alg√∫n otro gr√°fico que creas que te puedan pedir, como uno de pastel (pie chart) o algo as√≠? (Aunque casi no se usan en ciencia de datos, a veces los piden).
````

````python
¬°Claro que s√≠, mijo! Entiendo perfecto: quieres ver la diferencia clara entre c√≥mo se trabaja con **Matplotlib puro** y c√≥mo se trabaja con **Seaborn puro**.

Aqu√≠ te dejo los dos c√≥digos por separado para que veas la "personalidad" de cada librer√≠a.

---

### 1. El estilo Matplotlib (Control Manual y Tradicional)

Aqu√≠ t√∫ eres el arquitecto. Tienes que definir los colores, las leyendas y los ejes paso a paso. Se usa mucho para gr√°ficos personalizados.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos manuales (como tu ejercicio de los a√±os)
a√±os = np.arange(2000, 2011)
gdp_hawaii = [400, 420, 450, 480, 500, 530, 550, 580, 600, 620, 650]

# --- MATPLOTLIB PURO ---
plt.figure(figsize=(8, 5))

# Dibujamos la l√≠nea
plt.plot(a√±os, gdp_hawaii, color='red', linestyle='--', marker='o', label='Hawaii')

# Personalizaci√≥n manual de cada detalle
plt.title('Crecimiento GDP - Estilo Matplotlib', fontsize=14)
plt.xlabel('A√±o', fontsize=12)
plt.ylabel('GDP (Millones)', fontsize=12)
plt.grid(True, color='gray', alpha=0.3) # Cuadr√≠cula manual
plt.legend() # Obligatorio para que salga el label

# Anotaci√≥n manual
plt.annotate('Inicio de crisis', xy=(2008, 600), xytext=(2004, 620),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()

```

---

### 2. El estilo Seaborn (An√°lisis Estad√≠stico y Belleza Autom√°tica)

Aqu√≠ la librer√≠a hace el trabajo pesado. Solo le pasas el "DataFrame" y ella decide los colores, pone los nombres de los ejes solita y calcula la estad√≠stica.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos un dataset interno de Seaborn
df_tips = sns.load_dataset('tips')

# --- SEABORN PURO ---
# sns.set_theme() aplica un estilo global (fondo gris, cuadr√≠cula blanca)
sns.set_theme(style="darkgrid")

# Gr√°fico de viol√≠n: muestra distribuci√≥n y densidad de una sola vez
# (Muy dif√≠cil de hacer en Matplotlib puro, muy f√°cil en Seaborn)
plt.figure(figsize=(8, 5))
sns.violinplot(data=df_tips, x="day", y="total_bill", hue="sex", split=True, palette="Pastel1")

# Seaborn pone los t√≠tulos de los ejes autom√°ticamente bas√°ndose en las columnas
plt.title('Distribuci√≥n de Cuentas por D√≠a y G√©nero - Estilo Seaborn')
plt.show()

```

---

### üí° Diferencias clave para tu prueba:

| Caracter√≠stica | Matplotlib (El "Abuelo") | Seaborn (El "Moderno") |
| --- | --- | --- |
| **C√≥digo** | M√°s largo, l√≠nea por l√≠nea. | M√°s corto, basado en funciones de alto nivel. |
| **Est√©tica** | B√°sica, t√∫ tienes que ponerle colores. | Profesional y elegante por defecto. |
| **Datos** | Prefiere listas o arrays de Numpy. | Prefiere **DataFrames de Pandas**. |
| **Uso ideal** | Cuando necesitas poner una flecha o texto en un lugar muy espec√≠fico. | Cuando quieres ver tendencias, promedios y densidades r√°pido. |

**Un truco secreto:** En la vida real, casi siempre usamos los dos juntos. Usas **Seaborn** para crear el gr√°fico r√°pido y **Matplotlib** para poner el t√≠tulo final o guardar la imagen.

¬øQuieres que te prepare una pregunta "trampa" que suelen hacer los profes sobre cu√°ndo usar uno u otro?
````
