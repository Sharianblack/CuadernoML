## 6 frame.plot()

* (solo para Series) label: referencia para la leyenda
* ax: subfigura en la que dibujar los datos

* kind: tipo de gráfico (’bar’, ‘pie’, ‘hist’, ‘area’ ‘line’, ‘barh’, ‘density’, ‘kde’).
* xticks y yticks: valores en los ejes X e Y

```python
#scatter de clases, coloreando segun clase en matplotlib
iris = sns.load_dataset('iris')

# preparacion de los datos
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]
for i in groups:
    # busqueda por clase definida
    data.append(iris[iris["species"] == i][columns].values)

#creacion de la grafica
fig = plt.figure()
ax = fig.add_subplot()
for count, d in enumerate(data):
    # separar las variables del grafico
    x, y = d[:,0], d[:,1]
    ax.scatter(x, y, alpha = 0.4, c = colors[count -1], edgecolors = 'none', s = 30, label = groups[count - 1])
#etiquetas
ax.set_xlabel(columns[0])
ax.set_ylabel(columns[1])
plt.legend(loc='best')
plt.title('Matplot scatter plot')
plt.show()
```

## 8.1 matplotlib

1 crear una lista de años de 2000 a 2020 (eje x)
• generar datos aleatorios para el eje y (hacerlo 4 veces distintas: y0, y1, y2 e y3)
• Representar las 4 secuencias de datos aleatorios en una sola figura utilizando matplotlib
• Añade una leyenda para poder identificar cada secuencia en el gráfico
• Nombra las secuencias de la siguiente forma: “Hawaii”, “San Marino”, “Islas Feroe”,
“Guayana”
• Representar los mismos datos en 4 subfiguras (axes) distintas, parte de la misma figura
• Añade la siguiente info:
– Título: “Datos aleatorios”
– Leyenda
– Nombre del eje x: “Año”
– Nombre del eje y: “GDP”
• Añade al menos una anotación (texto y flecha) a uno de los gráficos

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Crear lista de años de 2000 a 2020 (Eje X)
# Usamos 2021 porque el límite superior es exclusivo

años = list(range(2000, 2021))

# 2. Generar datos aleatorios para el eje Y (4 secuencias distintas)
# Usamos un rango de 100 a 1000 para simular el GDP

np.random.seed(42) # Para que los datos no cambien cada vez que pruebes
y0 = np.random.randint(100, 1000, size=len(años)) # Hawaii
y1 = np.random.randint(100, 1000, size=len(años)) # San Marino
y2 = np.random.randint(100, 1000, size=len(años)) # Islas Feroe
y3 = np.random.randint(100, 1000, size=len(años)) # Guayana


# REPRESENTACIÓN EN UNA SOLA FIGURA (TODAS JUNTAS)

plt.figure(figsize=(10, 6))

plt.plot(años, y0, label="Hawaii", marker='o')
plt.plot(años, y1, label="San Marino", marker='s')
plt.plot(años, y2, label="Islas Feroe", marker='^')
plt.plot(años, y3, label="Guayana", marker='x')

# Añadir info
plt.title("Datos aleatorios - Comparativa GDP")
plt.xlabel("Año")
plt.ylabel("GDP")
plt.legend() # Muestra los nombres de las secuencias
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


# REPRESENTACIÓN EN 4 SUBFIGURAS (SUBPLOTS)

# Creamos una figura con una cuadrícula de 2 filas y 2 columnas
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Datos aleatorios", fontsize=16) # Título general

# Listas para facilitar el llenado de los subplots
datos = [y0, y1, y2, y3]
nombres = ["Hawaii", "San Marino", "Islas Feroe", "Guayana"]
colores = ['blue', 'orange', 'green', 'red']

# Llenamos cada subfigura usando un bucle
for i, ax in enumerate(axes.flat):
    ax.plot(años, datos[i], color=colores[i], label=nombres[i], linewidth=2)
    ax.set_title(f"Secuencia: {nombres[i]}")
    ax.set_xlabel("Año")
    ax.set_ylabel("GDP")
    ax.legend()
    ax.grid(True, alpha=0.3)


# AÑADIR ANOTACIÓN (TEXTO Y FLECHA)
# Vamos a señalar el punto más alto en el gráfico de Hawaii (el primero)

max_gdp = max(y0)
año_max = años[np.argmax(y0)]

# xy = donde apunta la flecha, xytext = donde va el texto
axes[0, 0].annotate('Punto más alto', 
                    xy=(año_max, max_gdp), 
                    xytext=(año_max + 1, max_gdp + 50),
                    arrowprops=dict(facecolor='black', shrink=0.05, width=2))

# Ajustar los espacios entre subplots para que no se amontonen
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 8.2 Seaborn

• Carga los datos de un dataset de seaborn (distinto a ‘tips’, ‘iris’ o ‘flights’)

• Si para alguno de los ejercicios no encuentras ejemplos en el dataset, busca otro para completarlo

• Representa los valores de dos variables (x e y, numéricas) en base a otra variable (categórica)


• Crea los gráficos que demuestren lo siguiente en el dataset seleccionado:
– La distribución de valores (media, distribución, outliers) de una variable numérica contínua
– Compara la distribución de esa variable con otra (contínua y numérica) para representar

– Muestra un ejemplo en el dataset de una variable a la que un modelo de regresión lineal
de orden > 1 sea más ajustado que uno de orden == 1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Configuramos el estilo de los gráficos
sns.set_theme(style="whitegrid")


# 1. CARGA DEL DATASET /  Usamos la URL raw como pide el ejercicio

url_raw = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/planets.csv"
df = pd.read_csv(url_raw)

print("--- Primeras filas del dataset ---")
print(df.head())


# 2. RELACIÓN DE CADA PAREJA DE ATRIBUTOS / El pairplot crea una matriz de gráficos comparando todas las variables numéricas

print("\nGenerando matriz de relaciones (pairplot)...")
sns.pairplot(df.dropna())  # dropna para evitar errores con valores nulos
plt.suptitle("Relación de Parejas de Atributos - Dataset Planets", y=1.02)
plt.show()


# 3. CORRELACIÓN DE UNA PAREJA (VALORES CONTINUOS Y NUMÉRICOS) / Escogemos 'orbital_period' y 'distance'

# Calculamos la matriz de correlación de Pearson
correlacion = df[['orbital_period', 'distance']].corr()
print("\n--- Correlación entre Periodo Orbital y Distancia ---")
print(correlacion)

# Demostración visual de la correlación con un heatmap pequeño o texto
plt.figure(figsize=(6, 4))
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title("Mapa de Calor: Correlación")
plt.show()


# 4. DOS VARIABLES NUMÉRICAS EN BASE A UNA CATEGÓRICA / X: distance, Y: orbital_period, HUE: method (categórica)

# Filtramos los 5 métodos más comunes para que el gráfico no sea un caos
top_methods = df['method'].value_counts().nlargest(5).index
df_filtered = df[df['method'].isin(top_methods)]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='distance', y='orbital_period', hue='method')
plt.yscale('log') # Escala logarítmica porque los periodos varían demasiado
plt.title("Distancia vs Periodo Orbital por Método de Descubrimiento")
plt.show()


# 5. DISTRIBUCIÓN DE UNA VARIABLE (Media, Distribución, Outliers) / Variable: 'distance'

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Boxplot para ver outliers
sns.boxplot(x=df['distance'], ax=axes[0], color='skyblue')
axes[0].set_title('Boxplot de Distancia (Identificación de Outliers)')

# Histplot para ver distribución y media
sns.histplot(df['distance'], kde=True, ax=axes[1], color='green')
axes[1].axvline(df['distance'].mean(), color='red', linestyle='--', label=f'Media: {df["distance"].mean():.2f}')
axes[1].set_title('Distribución de Distancia con Media')
axes[1].legend()
plt.show()


# 6. COMPARAR DISTRIBUCIÓN DE DOS VARIABLES (DENSIDAD) / Comparamos 'mass' y 'distance' usando un jointplot de tipo KDE

print("\nGenerando gráfico de densidad (jointplot)...")
# Eliminamos nulos para este gráfico específico
sns.jointplot(data=df.dropna(), x="distance", y="mass", kind="kde", fill=True, cmap='viridis')
plt.suptitle("Densidad Cuantitativa: Masa vs Distancia", y=1.02)
plt.show()

# 7. REGRESIÓN LINEAL: ORDEN 1 VS ORDEN > 1 /Vamos a ver cómo cambia la tendencia de descubrimiento por año

plt.figure(figsize=(10, 6))

# Filtramos datos para una mejor visualización (años recientes)

df_recent = df[df['year'] > 2000].dropna(subset=['distance'])

# Orden 1 (Lineal)

sns.regplot(data=df_recent, x='year', y='distance', order=1, 
            label='Orden 1 (Lineal)', scatter_kws={'alpha':0.3}, line_kws={'color':'red'})

# Orden 2 (Polinomial) - Suele ajustarse mejor si hay curvas en los datos

sns.regplot(data=df_recent, x='year', y='distance', order=2, 
            label='Orden 2 (Cuadrático)', scatter_kws={'alpha':0.3}, line_kws={'color':'blue'})

plt.title("Comparativa de Regresión: Año vs Distancia de Descubrimiento")
plt.legend()
plt.show()
```
